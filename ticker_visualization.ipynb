```python
import plotly
from plotly import express as px
import dash
import numpy as np
import pandas as pd
```

```python
full_data = pd.read_csv('stock_data.csv')
print( len( full_data.Name.unique() ) )
print( full_data.shape )
full_data.head()
```

```python
tickers = ['AAL', 'AAPL', 'AAP', 'ABBV', 'ABC', 'ABT', 'ACN', 'ADBE', 'ADI']
data = full_data[ full_data['Name'].isin( tickers ) ]
```

```python
fig = px.line(data, x='date', y = 'high', color = 'Name')
fig.show()
```

```python
options = { ticker: ticker for ticker in tickers }
app = dash.Dash('ticker_app')
app.layout = dash.html.Div( [ dash.dcc.Dropdown( options = options, value = tickers[0], multi=True, id = 'ticker_dropdown' ),
                              dash.dcc.Graph(id='ticker_chart', figure={}) ] )

@dash.callback( dash.Output('ticker_chart', 'figure'), dash.Input('ticker_dropdown', 'value') )
def _callback( value ):
    filterd_data  = data[ data['Name'].isin([value] if isinstance(value, str) else value ) ]
    fig = px.line(filterd_data, x = 'date', y = 'high', color='Name')
    return fig
```

```python
app.run(debug = True)
```
